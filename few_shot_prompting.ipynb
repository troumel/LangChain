```python
from openai import OpenAI

client = OpenAI()

examples = [
    {
        "role": "user",
        "content": "Βρες όλους τους πελάτες"
    },
    {
        "role": "assistant",
        "content": "SELECT * FROM customers;"
    },
    {
        "role": "user",
        "content": "Βρες όλα τα προϊόντα με τιμή πάνω από 100"
    },
    {
        "role": "assistant",
        "content": "SELECT * FROM products WHERE price > 100;"
    },
]

new_query = "Βρες όλους τους πελάτες από την Αθήνα"

messages = [
    {
        "role": "system",
        "content": (
            "You are an expert SQL assistant. "
            "Convert natural language queries into valid SQL statements. "
            "Return ONLY the SQL query, no explanations."
        )
    },
    *examples,
    {
        "role": "user",
        "content": new_query
    }
]

response = client.chat.completions.create(
    model="gpt-4o-mini",
    temperature=0,
    messages=messages
)

print(f"Input:  {new_query}")
print(f"Output: {response.choices[0].message.content}")
```
