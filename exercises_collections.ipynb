## Exercise 1: List Slicing Operations

Create a list of 5 numbers. Print the first 2, the last 2, and the reversed list using slicing.

```python
numbers = [10, 20, 30, 40, 50]

print("Original list:", numbers)
print("First 2 elements:", numbers[:2])
print("Last 2 elements:", numbers[-2:])
print("Reversed list:", numbers[::-1])
```

## Exercise 2: Dictionary Operations and Membership Checking

Create a dictionary with 3 people's names and ages. Check if a name exists using `in`, and use `.get()` with a default for a name that doesn't exist.

```python
people = {
    "Alice": 28,
    "Bob": 35,
    "Charlie": 42
}

print("Dictionary:", people)

print("\nChecking membership:")
print("'Alice' in people:", "Alice" in people)
print("'David' in people:", "David" in people)

print("\nUsing .get() method:")
print("Age of Alice:", people.get("Alice"))
print("Age of David (with default):", people.get("David", "Not found"))
```

## Exercise 3: Set Operations: Intersection and Difference

Create two sets of hobbies. Find the hobbies you share (intersection) and the ones only you have (difference).

```python
my_hobbies = {"reading", "gaming", "hiking", "cooking", "photography"}
friend_hobbies = {"gaming", "hiking", "drawing", "music", "cooking"}

print("My hobbies:", my_hobbies)
print("Friend's hobbies:", friend_hobbies)

# Find shared hobbies using intersection
shared_hobbies = my_hobbies.intersection(friend_hobbies)
print("\nShared hobbies (intersection):", shared_hobbies)

# Find hobbies only I have using difference
only_mine = my_hobbies.difference(friend_hobbies)
print("Hobbies only I have (difference):", only_mine)

# Find hobbies only friend has
only_friend = friend_hobbies.difference(my_hobbies)
print("Hobbies only friend has:", only_friend)
```

## Exercise 4: Sorting Lists with sorted() and .sort()

Create a list of numbers. Use `sorted()` to create a new sorted list in descending order, then use `.sort()` to sort the original list in ascending order.

```python
# Create a list of numbers
numbers = [45, 23, 78, 12, 89, 56, 34]

print("Original list:", numbers)

sorted_descending = sorted(numbers, reverse=True)
print("Sorted in descending order (using sorted()):", sorted_descending)
print("Original list after sorted():", numbers)

numbers.sort()
print("\nAfter using .sort() in ascending order:")
print("Original list (now sorted):", numbers)
```
