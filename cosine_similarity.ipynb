```python
import torch
import torch.nn.functional as F
```

```python
# Query: "Πού είναι τα λεφτά;"
# Έστω ότι το διάνυσμά του κοιτάει ψηλά και δεξιά
query_vec = torch.tensor([[0.8, 0.6]])
```

```python
# Documents (Η "Βάση Δεδομένων" μας)
# Doc A: "Η οικονομία πάει καλά" (Σχετικό) -> Κοιτάει προς τα εκεί
# Doc B: "Μου αρέσουν τα μήλα" (Άσχετο) -> Κοιτάει αλλού
# Doc C: "Τα χρήματα είναι στην τράπεζα" (Πολύ σχετικό) -> Κοιτάει σχεδόν ίδια
docs_vec = torch.tensor([
    [0.7, 0.7],  # Doc A
    [-0.9, 0.1], # Doc B (Αρνητικό X, κοιτάει αριστερά)
    [0.9, 0.4]   # Doc C
])

print(f"Query Shape: {query_vec.shape}")
print(f"Docs Shape:  {docs_vec.shape}")  
```

```python
# p=2 σημαίνει Ευκλείδειο μήκος (L2 norm)
query_norm = F.normalize(query_vec, p=2, dim=1)
docs_norm = F.normalize(docs_vec, p=2, dim=1)
print(f"Query Norm Shape: {query_norm.shape}") 
print(f"Docs Norm Shape:  {docs_norm.shape}") 
```

```python
scores = query_norm @ docs_norm.T

print(f"Similarity Scores: {scores}")
```

```python
# Βρες τα 2 καλύτερα (k=2)
# dim=1 γιατί ψάχνουμε στις στήλες των σκορ
top_scores, top_indices = torch.topk(scores, k=2, dim=1)

print("\n--- Search Results ---")
print(f"Indices: {top_indices}") # Ποια έγγραφα κέρδισαν;
print(f"Scores:  {top_scores}")  # Πόσο σκορ έπιασαν;
```
