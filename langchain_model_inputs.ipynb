```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.prompts import PromptTemplate
```

```python
chat = ChatOpenAI(model_name="gpt-4", model_kwargs={"seed": 365}, 
                   temperature=0,
                   max_tokens=100)
```

```python
response = chat.invoke('''I have recently adopted a dog. Can you suggest some names for him?''')
```

```python
print(response.content)
```

### Templates

```python
TEMPLATE = '''
System:
{description}

Human:
I have recently adopted a {pet}.
Can you suggest some {pet} names?
'''
```

```python
prompt_template = PromptTemplate.from_template(template = TEMPLATE)
prompt_template
```

```python
prompt_value = prompt_template.invoke({'description': 'You are a helpful assistant that suggests pet names in a sarcastical manner.', 'pet': 'cat'})
```

```python
print(prompt_value.text)
```
