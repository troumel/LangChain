## Creating a deployment checklist
Να δημιουργήσετε ένα prompt που ζητά ένα αναλυτικό checklist για deployment (π.χ. CI/CD,
environment variables, security checks)

```python
from openai import OpenAI
client = OpenAI()

prompt = """
You are a senior DevOps engineer with 10+ years of experience managing CI/CD pipelines
and production deployments across cloud and on-premise environments.

A development team is preparing to deploy a new web application to production for the
first time. They need a thorough pre-deployment review to avoid downtime, security
vulnerabilities, and configuration errors.

Create a comprehensive deployment checklist that covers all critical areas a team must
verify before going live. Include at minimum:
- CI/CD pipeline setup and health
- Environment variables and secrets management
- Security checks
- Database migrations
- Monitoring and alerting
- Rollback strategy

Return the checklist as clearly labelled sections (using headers like ## Section Name).
Under each section, list actionable items as checkboxes (- [ ] item).
Be specific and practical — avoid generic advice.
"""

response = client.chat.completions.create(
    model="gpt-4o-mini",
    temperature=0,
    messages=[
        {"role": "user", "content": prompt}
    ]
)

print(response.choices[0].message.content)
```
