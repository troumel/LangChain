```python
import numpy as np
```

```python
data = np.array([1,2])
print(data)
```

```python
ones = np.ones(2)
ones
```

```python
zeros = np.zeros(3)
zeros
```

```python
random = np.random.rand(3)
random
```

```python
print(data + ones)
print(data * data)
print(data - ones)
print(data / data)
print(data * 1.6) #broadcasting example 
```

```python
#indexing
data_3 = np.array([1,2,3])
print(data_3[0])  # First element
print(data_3[1])  # Second element
print(data_3[2])  # Third element
print(data_3[-1]) # Last element
print(data_3[-2]) # Second last element
print(data_3[-3]) # Third last element
print(data_3[:2]) # Slicing: first two elements or print(data_3[0:2])
print(data_3[1:3]) # Slicing: last two elements or print(data_3[1:])
```

```python
print(data_3.max())  # Maximum value
print(data_3.min())  # Minimum value
print(data_3.sum())  # Sum of all elements
print(data_3.mean()) # Mean of all elements
print(data_3.std())  # Standard deviation
```

```python
print(np.array([[1,2,3],[4,5,6]]))  # 2D array
```

```python
print(np.ones((2,3)))  # 2D array of ones
print(np.zeros((3,2)))  # 2D array of zeros
```

```python
print(np.ones((1,3)) * np.zeros((2,3)))  # Element-wise multiplication of 2D arrays
```
