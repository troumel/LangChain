```python
# Matrix multiplication (correct way)
A = np.ones((2,3))  # 2x3 matrix
B = np.zeros((3,2))  # 3x2 matrix

print("Matrix A (2x3):")
print(A)
print("\nMatrix B (3x2):")
print(B)

# Matrix multiplication using @ operator
print("\nMatrix multiplication A @ B:")
print(A @ B)

# Alternative ways to do matrix multiplication:
print("\nUsing np.matmul():")
print(np.matmul(A, B))

print("\nUsing np.dot():")
print(np.dot(A, B))
```

```python
# Demonstrating the difference between * and @ operators
C = np.array([[1, 2], [3, 4]])
D = np.array([[5, 6], [7, 8]])

print("Matrix C:")
print(C)
print("\nMatrix D:")
print(D)

print("\nElement-wise multiplication (C * D):")
print(C * D)

print("\nMatrix multiplication (C @ D):")
print(C @ D)
```

Email excercise / urgent vs non-urgent


```python
import torch.nn.functional as F
```

```python
# Σκληροπυρηνικά δεδομένα για το παράδειγμα (Logits)
# Email A: Πολύ μεγάλο σκορ στο "Ασήμαντο" (4.5 vs -2.0)
# Email B: Μπερδεμένο, λίγο προς το "Επείγον" (0.1 vs 0.2)
# Email C: Πολύ μεγάλο σκορ στο "Επείγον" (-3.0 vs 5.0)
logits = torch.tensor([
    [ 4.5, -2.0],  # Email A (ID: 101)
    [ 0.1,  0.2],  # Email B (ID: 102)
    [-3.0,  5.0]   # Email C (ID: 103)
])

print(f"Shape: {logits.shape}")
```

### apply softmax fro logit probabilites

```python
probs = F.softmax(logits, dim=1)
print(probs)
```

```python
urgent_scores = probs[:, 1]

print(f"Urgent Scores: {urgent_scores}")
print(f"Shape: {urgent_scores.shape}")
```

### threshold filtering/masking


```python
threshold = 0.9

# create true false mask
is_urgent = urgent_scores > threshold

print(f"Is Urgent Mask: {is_urgent}")
```

### retrieval

```python
# Βρες τις θέσεις που είναι True
indices = is_urgent.nonzero()

print(f"Indices: {indices}")
# Output: tensor([[2]]) -> Το email στη θέση 2 (το 3ο email δηλαδή)


# Μετατροπή σε απλή λίστα Python
urgent_email_indices = indices.squeeze().tolist() 
print(urgent_email_indices)

# Χειρισμός αν είναι ένας αριθμός ή λίστα
if isinstance(urgent_email_indices, int):
    urgent_email_indices = [urgent_email_indices]

print(f"Send these indices to DB: {urgent_email_indices}")
```

### Padding/attention mask

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

emails = [
    "Hello",                   # Μικρό email
    "Please help me right now" # Μεγάλο email
]

# Ο Tokenizer το κάνει αυτόματα για σένα με το padding=True
inputs = tokenizer(emails, padding=True, return_tensors="pt")

print("Input IDs (Τα δεδομένα):")
print(inputs['input_ids'])

print("\nAttention Mask (Ο οδηγός):")
print(inputs['attention_mask'])
```
