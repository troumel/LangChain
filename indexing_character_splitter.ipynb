```python
from langchain_community.document_loaders import Docx2txtLoader
from langchain_text_splitters.character import CharacterTextSplitter
```

```python
loader = Docx2txtLoader("Introduction_to_Data_and_Data_Science.docx")
pages = loader.load()
print(pages)

for i in range(len(pages)):
    pages[i].page_content = ' '.join(pages[i].page_content.split())
```

```python
pages[0].page_content
```

```python
len(pages[0].page_content)
```

```python
character_splitter = CharacterTextSplitter(
    separator= "",
    chunk_size= 500,
    chunk_overlap= 200,
)
```

```python
pages_character_split = character_splitter.split_documents(pages)
pages_character_split
```

```python
print(len(pages_character_split))
print(len(pages_character_split[0].page_content))
print(len(pages_character_split[16].page_content))
```

### dot separation

```python
character_splitter = CharacterTextSplitter(
    separator= ".",
    chunk_size= 500,
    chunk_overlap= 200,
)
```

```python
pages_character_split = character_splitter.split_documents(pages)
print(pages_character_split)
print(len(pages_character_split))
print(len(pages_character_split[0].page_content))
print(len(pages_character_split[16].page_content))
```
