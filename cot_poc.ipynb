```python
from openai import OpenAI
client = OpenAI()
```

```python
def process_numbers(nums):
    result = []
    for num in nums:
        if num % 2 == 0:
            result.append(num * 2)
        else:
            result.append(num * 3)
    return sum(result)

numbers = [1, 2, 3, 4]
```

```python
prompt_direct = """Given this function and input, return only the final numeric result.\n\n
def process_numbers(nums):\n
    result = []\n
    for num in nums:\n
        if num % 2 == 0:\n
            result.append(num * 2)\n
        else:\n
            result.append(num * 3)\n
    return sum(result)\n\n
numbers = [1, 2, 3, 4]\n\n
What is process_numbers(numbers)?\n
Answer with only the number."""

prompt_cot = """Given this function and input, think step by step.
Explain what happens in each loop iteration (value of num, branch chosen, value appended),
then compute the final sum and give the final numeric result at the end.\n\n
def process_numbers(nums):\n
    result = []\n
    for num in nums:\n
        if num % 2 == 0:\n
            result.append(num * 2)\n
        else:\n
            result.append(num * 3)\n
    return sum(result)\n\n
numbers = [1, 2, 3, 4]\n\n
What is process_numbers(numbers)?"""

direct_response = client.chat.completions.create(
    model="gpt-4o-mini",
    temperature=0,
    messages=[
        {"role": "user", "content": prompt_direct}
    ]
)

cot_response = client.chat.completions.create(
    model="gpt-4o-mini",
    temperature=0,
    messages=[
        {"role": "user", "content": prompt_cot}
    ]
)

print("Direct prompt answer:")
print(direct_response.choices[0].message.content)
print()
print("CoT prompt answer:")
print(cot_response.choices[0].message.content)
```
