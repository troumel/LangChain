## Prompt A (bad prompt)
- Explain APIs

## Prompt B (so-so prompt)
- Explain what API is and give example

## Prompt C (good prompt)
- You are a software architect. Explain what an API is to non-technical business
stakeholders. Use a real-world analogy and provide one practical example from web
development. Structure your answer in bullet points.

## Prompt Evaluation

Each prompt is graded **1–5** across five characteristics:

| Characteristic | Description |
|----------------|-------------|
| **Clarity** | Is the request unambiguous and precise? |
| **Context** | Is there enough background for the model to tailor the response? |
| **Persona** | Is there a role/expertise assigned to the model? |
| **Output Quality** | How predictable, useful, and accurate is the expected response? |
| **Format** | Is the desired structure/length of the answer specified? |

---

### Prompt A — `"Explain APIs"`

| Characteristic | Score | Justification |
|----------------|-------|---------------|
| Clarity | 1/5 | Completely open-ended — "explain" in what way, at what depth, for whom? |
| Context | 1/5 | No context at all; the model has to guess everything |
| Persona | 1/5 | No role assigned; the model will respond as a generic assistant |
| Output Quality | 1/5 | Response will be unpredictably long, generic, and not targeted |
| Format | 1/5 | No structure requested; likely returns a wall of text |

**Overall: 1/5** — A minimal prompt that will produce a technically correct but unfocused answer.

---

### Prompt B — `"Explain what API is and give example"`

| Characteristic | Score | Justification |
|----------------|-------|---------------|
| Clarity | 3/5 | Slightly clearer — requests an example, but audience and depth are still undefined |
| Context | 2/5 | No context about who is asking or why |
| Persona | 1/5 | Still no persona |
| Output Quality | 2/5 | Will get a basic definition + one example, but style and depth remain unpredictable |
| Format | 1/5 | No format instruction; structure of the answer is left to the model |

**Overall: 2/5** — A mild improvement over A, but still leaves too much to chance.

---

### Prompt C — Full PCTF-style prompt

| Characteristic | Score | Justification |
|----------------|-------|---------------|
| Clarity | 5/5 | Precise audience (non-technical stakeholders), specific requirements (analogy + web example) |
| Context | 4/5 | Audience and use-case are clear; slightly deducted as no industry/company context |
| Persona | 5/5 | Explicit role ("software architect") shapes tone and vocabulary perfectly |
| Output Quality | 5/5 | Constrained enough to produce a consistent, actionable, audience-appropriate response |
| Format | 5/5 | Explicitly requests bullet points |

**Overall: 5/5** — A well-engineered prompt that will reliably produce targeted, structured output.

```python
from openai import OpenAI
client = OpenAI()

prompt_a = """
Explain APIs
"""

prompt_b = """
Explain what API is and give example
"""

prompt_c = """
You are a software architect. Explain what an API is to non-technical business
stakeholders. Use a real-world analogy and provide one practical example from web
development. Structure your answer in bullet points.
"""

response_a = client.chat.completions.create(
    model="gpt-4o-mini",
    temperature=0,
    messages=[
        {"role": "user", "content": prompt_a}
    ]
)

response_b = client.chat.completions.create(
    model="gpt-4o-mini",
    temperature=0,
    messages=[
        {"role": "user", "content": prompt_b}
    ]
)

response_c = client.chat.completions.create(
    model="gpt-4o-mini",
    temperature=0,
    messages=[
        {"role": "user", "content": prompt_c}
    ]
)

print("=== Prompt A (bad) ===")
print(response_a.choices[0].message.content)
print()
print("=== Prompt B (so-so) ===")
print(response_b.choices[0].message.content)
print()
print("=== Prompt C (good) ===")
print(response_c.choices[0].message.content)
```
