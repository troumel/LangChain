## Exercise 1: List Comprehension for Cubes

Use a list comprehension to create a list of the first 10 cubes (1³, 2³, ..., 10³)

```python
cubes = [x**3 for x in range(1, 11)]

print("List of first 10 cubes:")
print(cubes)

print("\nBreakdown:")
for i, cube in enumerate(cubes, 1):
    print(f"{i}³ = {cube}")
```

## Exercise 2: List Comprehension with Filter

Given `words = ["Hello", "", "World", "", "Python"]`, use a list comprehension to keep only non-empty strings and convert them to lowercase

```python
words = ["Hello", "", "World", "", "Python"]

print("Original words:", words)

filtered_words = [word.lower() for word in words if word]

print("Filtered and lowercase:", filtered_words)

filtered_words_alt = [word.lower() for word in words if word != ""]
print("Alternative (explicit condition):", filtered_words_alt)
```

## Exercise 3: Dictionary Comprehension to Invert Keys and Values

Use a dict comprehension to invert a dictionary (swap keys and values). Given `{"a": 1, "b": 2, "c": 3}`, produce `{1: "a", 2: "b", 3: "c"}`

```python
original_dict = {"a": 1, "b": 2, "c": 3}

print("Original dictionary:", original_dict)

inverted_dict = {value: key for key, value in original_dict.items()}

print("Inverted dictionary:", inverted_dict)

print("\nVerification:")
for key, value in inverted_dict.items():
    print(f"  {key}: {value}")
```

## Exercise 4: List Comprehension with if/else

Use a list comprehension with if/else to replace all negative numbers with "negative" and keep positives as-is. Input: `[-5, 3, -1, 7, 0, -2]`

```python
numbers = [-5, 3, -1, 7, 0, -2]

print("Original list:", numbers)

result = ["negative" if x < 0 else x for x in numbers]

print("Result (negatives replaced):", result)

result_no_negatives = [x if x >= 0 else "negative" for x in numbers]
print("Alternative (same result):", result_no_negatives)

print("\nBreakdown:")
for original, transformed in zip(numbers, result):
    print(f"  {original:3} -> {transformed}")
```
